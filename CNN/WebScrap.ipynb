```python
import requests
from bs4 import BeautifulSoup as BS
from selenium import webdriver
import time
import pandas as pd
```

```python
url = 'https://www.suffolk.edu/academics/faculty/'
```

```python
driver = webdriver.Chrome(executable_path="C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
```

```python
driver.get(url)
```

```python
time.sleep(3)
```

```python
html = driver.page_source
```

```python
ht = BS(html,"html.parser")
```

```python
Facultylist=ht.find("div",{"class":"faculty-listing faculty-list-page"})
```

```python
Facultylist
```

```python
Facultylinks = []
```

```python
possible_links = Facultylist.find_all('a', {"class": "faculty-item"})
for link in possible_links:
    if link.has_attr('href'):
        Facultylinks.append(link.attrs['href'])
```

```python
len(Facultylinks)
```

```python
FacultyDetails = []
```

```python
for url in Facultylinks:
    ht = BS(requests.get("https://www.suffolk.edu"+url).text, "html.parser")
    time.sleep(2)
    name = ht.find("h1", {"class": "title"}).text
    post = ht.find("p", {"class": "sub-title"})
    facultyBox = ht.find("div", {"class": "faculty-box"})
    phone = facultyBox.find("li").text
    email = facultyBox.find("a").text
    if post is None:
        post = "Unavailable"
    else:
        post = post.text
    FacultyDetails.append({"Name": name, "Designation": post, "Phone Number": phone, "Email": email, "URL":
        "https://www.suffolk.edu"+url})
```

```python
rLinks=Facultylinks[415:]
```

```python
rLinks
```

```python
for url in rLinks:
    ht = BS(requests.get("https://www.suffolk.edu"+url).text, "html.parser")
    time.sleep(2)
    name = ht.find("h1", {"class": "title"}).text
    post = ht.find("p", {"class": "sub-title"})
    facultyBox = ht.find("div", {"class": "faculty-box"})
    phone = facultyBox.find("li").text
    email = facultyBox.find("a").text
    if post is None:
        post = "Unavailable"
    else:
        post = post.text
    FacultyDetails.append({"Name": name, "Designation": post, "Phone Number": phone, "Email": email, "URL":
        "https://www.suffolk.edu"+url})
```

```python
len(FacultyDetails)
```

```python
df=pd.DataFrame(FacultyDetails)
```

```python
col = ['Name','Designation','Phone Number','Email','URL']
```

```python
df = df.reindex(columns=col)
```

```python
df
```

```python
NameList=df.Name.tolist()
```

```python
def splitName(li):
    NameQual=[]
    for i in li:
        temp = i.split(',')
        if not len(temp)==2:
            temp.append("Unavailable")
        NameQual.append({"Name":temp[0],"Q":temp[1]})
    return NameQual
```

```python
d=splitName(NameList)
```

```python
d
```

```python
d[1]['Name']
```

```python
df.Name[1]=d[1]['Name']
```

```python
df.Name[1]
```

```python
for i in range(0,len(d)):
    if d[i]['Q'] == ' PhD':
        st = "Dr "+d[i]['Name']
    else:
        st = d[i]['Name']
    df.Name[i]=st
```

```python
df
```

```python
PostDept=df.Designation.tolist()
```

```python
PostDept[1].split(',')
```

```python
def splitPost(li):
    PostDept=[]
    for i in li:
        temp = i.split(',')
        if not len(temp)==2:
            temp.append("Unavailable")
        PostDept.append({"P":temp[0].strip(),"D":temp[1].strip()})
    return PostDept
```

```python
PoDe=splitPost(PostDept)
```

```python
jj=[]
for i in range(len(PoDe)):
    jj.append(PoDe[i]['P'])
```

```python
jj
```

```python
df['Post']=jj
```

```python
df
```

```python
DeptList=[]
for i in range(len(PoDe)):
    DeptList.append(PoDe[i]['D'])
```

```python
for i in range(len(DeptList)):
    if len(DeptList[i]) > 30:
        DeptList[i]=DeptList[i].split(';')[0]
```

```python
for i in range(len(DeptList)):
    if len(DeptList[i]) > 30:
        DeptList[i]=DeptList[i][:30]
```

```python
df['Dept']=DeptList
```
